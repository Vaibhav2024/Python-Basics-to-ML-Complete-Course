# Data Visulaization using seaborn

seaborn is python visulaization library based on matplotlib that provides a high level interface for drawing attractive and informative statistical graphics. Seaborn hepls in creating complex visulaization with just a few line of code.

```python
import seaborn as sns
```

### Basic Plotting with Seaborn

```python
tips = sns.load_dataset('tips')
tips
```

## creating a scatter plot

```python
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title("Scatter plot of total bill vs tips")
plt.show()
```

## Line Plot

```python
sns.lineplot(x="size", y="total_bill", data=tips)
plt.title("Line plot of total bill by size")
plt.show()
```

## Categorical Plots


```python
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("Bar Plot of total bill by day")
plt.show()
```

## Box Plot

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
# the circle are outliers and the box represent 25,50,75% form
```

## Violin Plot

```python
sns.violinplot(x="day", y="total_bill", data=tips)
plt.show()
```

## Histogram

```python
sns.histplot(tips['total_bill'], bins=10, kde=True) # kde means kernel density estimator,
                                                    # shows a line of data flow if true
```

## Kernel Density Estimator (KDE) Plot

```python
sns.kdeplot(tips['total_bill'],fill=True)      #fill=True means it fill the blue color
```

## Pair Plot

```python
sns.pairplot(tips)
# shows relation between every column with numeric value
```

## Heatmap

```python
corr = tips[['size', 'total_bill', 'tip']].corr()
print(corr)
sns.heatmap(corr, annot=True, cmap='coolwarm')  #annot=True annot value in heat map, 
#cmap=" " are color pelets for heatmap, find others in seaborn documentation
```

# Practical 

```python
import pandas as pd
sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
## Total Sales by product
plt.figure(figsize=(10, 6))                         # in seaborn we use estimator instead of writing sum()
sns.barplot(x="Product Category", y='Total Revenue', data=sales_df, estimator=sum, color='purple')
plt.title("Total Sales of Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show() 
```
