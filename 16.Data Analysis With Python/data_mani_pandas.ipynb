```python
import pandas as pd
df = pd.read_csv("data.csv")
```

```python
df.head(5)
```

#### Handling Missing Values  , isnull(), .any(), .sum()

```python
print(df.head(5).isnull())     #used to see null value in true or false format
print()
print(df.isnull().any(axis=0))       #used to get value with respect to whole column or row using axis
print()
print(df.isnull().sum())        #used to get total number of null value in a column
```

## filling missing value with mean of column, fillna()

```python
dp1 = df.fillna(0)    #fillna is used to fill null value but remeber to use inplace to make it permenant
print(dp1.head(5))
```

```python
# filling missing value with mean of the column
df["Sales_fillNA"] = df["Sales"].fillna(df['Sales'].mean())
df.head(15)
```

### Reanmaing Columns, df.rename(columns={"":""})

```python
df2 = df.rename(columns={'Date': 'Sales Date'})
df2.head()
```

### Changing Datatype,  astype(float)

```python
df["Value_New"] = df['Value'].fillna(df["Value"].mean()).astype(int)    #astype() is used to convert data type
print(df.head())
print(df.dtypes)
```

### Applying certain operation to column values, apply()

```python
df['New Value'] = df['Value'].apply(lambda x:x*2)
df.head(5)
```

## Data Aggregating and Grouping,   groupby()

```python
# groupby basicaly groudps by category and value of column
grouped_mean = df.groupby('Product')['Value'].mean()
print(grouped_mean)
```

```python
# groupby using multiple category
grouped_mean2 = df.groupby(['Product', 'Region'])['Value'].mean()
print(grouped_mean2)
# here we are raking mean by grouping product and ther region to find mean
```

```python
# using group by for sum
grouped_sum = df.groupby(['Product', 'Region'])['Value'].sum()
print(grouped_sum)
```

### Aggreating multiple function, aggregating means appling multiple build in math methods to particular category, and there value

```python
grouped_agg = df.groupby(['Product', 'Region'])['Value'].agg(['mean', 'sum', 'count', 'median'])
grouped_agg
```

## Merging and Joining DataFrames, on="Key", how="inner, outer,"

```python
df1 = pd.DataFrame({'Key': ['A', 'B', 'C'], 'Value1': [1, 2, 3]})
df2 = pd.DataFrame({'Key': ['A', 'B', 'D'], 'Value2': [4, 5, 6]})

df3 = pd.merge(df1, df2, on="Key", how="inner") #inner gives the value that are similar in both key
print(df3)  #on: on which common key merging should take place
print()

df4 = pd.merge(df1, df2, on="Key", how="outer") #outer is used for all keys, even one is not present in
print(df4)  # another if so it will print NAN in those places
print()

df5 = pd.merge(df1, df2, on='Key', how="left")  #gives more priority to left dataframe, only takes common
print(df5)  #value present in dataframe1 and all value of dataframe 1 exclude the non common from dataframe2

df6 = pd.merge(df1, df2, on='Key', how="right")  #gives more priority to right dataframe, only takes common
print(df5)  #value present in dataframe2 and all value of dataframe2 exclude the non common from dataframe1
```
